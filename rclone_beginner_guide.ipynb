This guide uses **simple English** and avoids unnecessary theory.

---

## What is `rclone`?

`rclone` is a command-line tool for working with cloud storage.

You can use it to:
- upload files
- download files
- copy folders
- sync folders
- list files
- check storage
- move data between your computer and cloud drives

Think of it like this:

- **your computer folder** = one storage place
- **Google Drive / OneDrive / Dropbox** = another storage place
- **rclone** = the delivery worker that moves files between them

---

## The most important idea first

With `rclone`, you usually work with **two sides**:

1. **local path** = a folder on your computer  
2. **remote path** = a folder on cloud storage

Example idea:

- Local: `/home/you/project`
- Remote: `gdrive:backup/project`

That means:
- source can be local or remote
- destination can be local or remote

---

## Very important word: `remote`

In `rclone`, a **remote** is just a nickname for a cloud account.

Examples:
- `gdrive` = your Google Drive
- `onedrive` = your Microsoft OneDrive
- `dropbox` = your Dropbox

So when you see:

```bash
gdrive:myfolder
```

it means:

- use the remote named `gdrive`
- go to the folder `myfolder` inside that cloud storage

Authentication means:
- proving that this cloud account is yours
- giving `rclone` permission to access your files

Usually this happens through a browser login.


# 9. `copy` — the safest command for beginners

If you learn only one transfer command first, learn `copy`.

```bash
rclone copy /home/you/data gdrive:backup/data -P
```

## What `copy` does
Copies files from source to destination.

Important:
- it copies new or changed files
- it **does not delete** extra files from destination

That is why it is safer than `sync` for beginners.

---

## Command breakdown

```bash
rclone copy /home/you/data gdrive:backup/data -P
```

### Part by part

- `rclone`  
  the program

- `copy`  
  the command

- `/home/you/data`  
  the **source**  
  this is your local folder

- `gdrive:backup/data`  
  the **destination**  
  this is your Google Drive folder

- `-P`  
  show progress

---

## Very important rule

`copy` is good when you want:
- upload without deleting destination files
- download without deleting local files
- safer first tests

---

## Example: upload to OneDrive

```bash
rclone copy /home/you/results onedrive:results -P
```

## Example: upload to Dropbox

```bash
rclone copy /home/you/report dropbox:report -P
```

## Example: download from Google Drive to local computer

```bash
rclone copy gdrive:dataset /home/you/dataset -P
```


# 10. `copyto` — copy one file to one exact destination name

Use `copyto` when you want to copy **one file** and control the exact destination file name.

```bash
rclone copyto /home/you/report.pdf gdrive:docs/report.pdf -P
```

### Meaning
- source = one exact file
- destination = one exact file path

### Why use `copyto`?
Because sometimes `copy` is for folder-to-folder thinking.

`copyto` is easier when your brain says:
> "Take this file and put it exactly there."

---

## Example
```bash
rclone copyto gdrive:docs/report.pdf /home/you/Downloads/report.pdf -P
```

This downloads one file from Google Drive to local.


# 11. `sync` — powerful, useful, dangerous

```bash
rclone sync /home/you/data gdrive:backup/data -P
```

## What `sync` does
Makes the **destination** look like the **source**.

That means:
- new files are copied
- changed files are updated
- files missing from source can be deleted from destination

---

## The most important beginner warning

`sync` can delete files from the destination.

So:

```bash
rclone sync A B
```

means:

- **A = source (truth)**
- **B = destination (will be changed to match A)**

---

## Easy memory trick

> `sync SOURCE DESTINATION`  
> destination becomes like source

---

## When to use `sync`
Use it when you really want a mirror.

Examples:
- keep backup folder exactly the same
- make remote folder match local folder
- repeated backup job

---

## When NOT to use `sync`
Do not use it when:
- you are unsure
- you are afraid of deleting files
- you just want to upload extra files safely

In those cases, use `copy`.

---

## Beginner safety habit

Always test with dry run first:

```bash
rclone sync /home/you/data gdrive:backup/data --dry-run -P
```

This shows what would happen **without actually changing files**.


# 12. `move` — copy then remove from source

```bash
rclone move /home/you/finished gdrive:archive/finished -P
```

## What `move` does
Moves files from source to destination.

That means:
- data is transferred
- after successful transfer, source files are removed

### Use case
You want to archive completed data and remove it from the original place.

### Beginner warning
Do not use `move` unless you are sure.

If you want a safer start, use `copy` first.


# 13. `delete` — remove files only

```bash
rclone delete gdrive:old-data
```

## What it does
Deletes files in the given path.

### Important
This removes files.

Use carefully.

---

## Beginner advice
Prefer to:
1. list files first
2. check path carefully
3. avoid delete until you are sure

Because cloud mistakes are annoying.

# 16. The arguments and flags you really need to understand

These are the most useful flags for beginners.

---

## `-P` or `--progress`

Example:
```bash
rclone copy /home/you/data gdrive:backup -P
```

### Meaning
Show transfer progress in real time.

### Why it helps
You can see:
- speed
- transferred amount
- estimated progress

For beginners, this is almost always useful.

---

## `--dry-run`

Example:
```bash
rclone sync /home/you/data gdrive:backup --dry-run -P
```

### Meaning
Show what would happen, but do not actually change anything.

### Why it helps
This is the best safety tool.

Use it before:
- `sync`
- `move`
- `delete`
- any scary command

---

## `--interactive` or `-i`

### Meaning
Ask before making important changes.

### Why it helps
Extra safety when you are nervous.

---

## `--transfers`

Example:
```bash
rclone copy /home/you/data gdrive:backup -P --transfers 4
```

### Meaning
How many file transfers happen in parallel.

### Simple explanation
Higher number can be faster, but can also use more bandwidth or API requests.

### Beginner advice
Start small:
- `--transfers 2`
- `--transfers 4`

Do not start by making everything huge.

---

## `--checkers`

Example:
```bash
rclone copy /home/you/data gdrive:backup -P --checkers 8
```

### Meaning
How many checking operations run in parallel.

### Simple explanation
These workers help inspect files and compare information.

### Beginner advice
Usually you do not need to change this early.
But if you do, common small values are fine.

---

## `--ignore-existing`

Example:
```bash
rclone copy /home/you/data gdrive:backup --ignore-existing -P
```

### Meaning
Skip files that already exist on destination.

### Use case
You want to avoid touching files already there.

---

## `--update`

Example:
```bash
rclone copy /home/you/data gdrive:backup --update -P
```

### Meaning
Skip files that are newer on the destination.

### Use case
Helpful when destination might already have newer versions.

---

## `--create-empty-src-dirs`

### Meaning
Also create empty folders from the source.

### Why this matters
Normally transfer commands care mostly about files.
This flag helps preserve empty folder structure too.

---

## `--max-size`

Example:
```bash
rclone copy /home/you/data gdrive:backup --max-size 100M -P
```

### Meaning
Copy only files up to that maximum size.

### Use case
You want to exclude very large files.

---

## `--min-size`

### Meaning
Copy only files above a minimum size.

---

## `--exclude`

Example:
```bash
rclone copy /home/you/data gdrive:backup --exclude "*.tmp" -P
```

### Meaning
Ignore files matching a pattern.

### Use case
Skip temporary files, cache files, junk files.

---

## `--include`

### Meaning
Only include files matching a pattern.

Example idea:
- only `.fastq.gz`
- only `.pdf`
- only `.csv`

---

## `--log-file`

Example:
```bash
rclone copy /home/you/data gdrive:backup -P --log-file rclone.log
```

### Meaning
Save log output to a file.

### Use case
Very useful for long jobs.

---

## `--stats`

Example:
```bash
rclone copy /home/you/data gdrive:backup -P --stats 10s
```

### Meaning
Show stats every given time interval.

### Example meaning
`10s` = every 10 seconds


# 17. How to think about SOURCE and DESTINATION

This part causes many beginner mistakes.

## Rule:
```bash
rclone COMMAND SOURCE DESTINATION
```

Examples:

### Upload local -> cloud
```bash
rclone copy /home/you/data gdrive:backup/data -P
```

### Download cloud -> local
```bash
rclone copy gdrive:backup/data /home/you/data -P
```

### Cloud -> cloud
```bash
rclone copy gdrive:backup onedrive:backup -P
```

---

## Easy memory trick

- left side = where files come from
- right side = where files go to

---

## Workflow C: download one folder safely

```bash
rclone copy gdrive:dataset /home/you/dataset --dry-run -P
rclone copy gdrive:dataset /home/you/dataset -P
```

---

## Workflow D: mirror backup carefully

```bash
rclone sync /home/you/data gdrive:mirror/data --dry-run -P
rclone sync /home/you/data gdrive:mirror/data -P
```

### Why the dry run matters
Because `sync` can delete destination files.

```bash
rclone lsd gdrive:
rclone lsl gdrive:backup
```

# 24. The most common beginner mistakes

## Mistake 1: confusing `copy` and `sync`

### `copy`
- safer
- does not delete extra files in destination

### `sync`
- stronger
- destination becomes like source
- may delete files from destination

If confused, use `copy` first.

---

## Mistake 2: reversing source and destination

Bad thinking:
> "I know what I mean in my head."

No. Check carefully.

Always read the command as:
```text
from SOURCE to DESTINATION
```

---

## Mistake 3: not using `--dry-run`

If the command feels risky:
- use `--dry-run`

Especially for:
- `sync`
- `move`
- `delete`

---

## Mistake 4: not checking the remote first

Before upload/download:
```bash
rclone listremotes
rclone lsd gdrive:
```

Make sure the remote actually works.

---

## Mistake 5: doing a huge transfer first

Do not start with 600 GB on day one.

Start with:
- one tiny file
- one small folder
- then bigger jobs

---

## Mistake 6: bad remote names

Use:
- `gdrive`
- `onedrive`
- `dropbox`

Avoid nonsense names.
